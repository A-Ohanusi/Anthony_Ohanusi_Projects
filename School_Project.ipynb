```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


S_Survey = pd.read_excel('Do Now Survey.xlsx')
S_Survey.head(5)
```

```python
print(S_Survey.columns.tolist)
```

```python
Survey_data = S_Survey[['Id'
, 'How often does the "Do Now" activity help you settle into lessons quickly and quietly at the start of class? \n '
, 'Compared to last term, how often do lessons now start without disruptions (e.g., talking or off-task behavior)?  '
,  'How often do you find the "Do Now" activity (5 questions in 5 minutes) engaging or interesting?  '
, 'How often do you feel that lessons this term start on time compared to last term? \n '
,'How often do you feel that staff on duty in corridors make transitions between lessons smoother (e.g., less pushing or crowding)?  '
, 'Overall, how often do you feel the start of lessons this term is more organized and disruption-free compared to last term?  ']]

Survey_data.head()
```

```python
Summary1 =Survey_data['How often does the "Do Now" activity help you settle into lessons quickly and quietly at the start of class? \n '].value_counts().reset_index()
Summary1.columns = ['How often does the "Do Now" activity help you settle into lessons quickly and quietly at the start of class?', 'Count']
Summary1
```

```python
labels = Summary1['How often does the "Do Now" activity help you settle into lessons quickly and quietly at the start of class?']
counts = Summary1['Count']
colors = plt.cm.rainbow(np.linspace(0, 1, len(labels)))

# Plot
plt.figure(figsize=(10, 6))
plt.bar(labels, counts)
plt.title('How often does the "Do Now" activity help you settle into lessons quickly and quietly at the start of class?')
plt.xlabel('Response')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
Summary2 = Survey_data['Compared to last term, how often do lessons now start without disruptions (e.g., talking or off-task behavior)?  '].value_counts().reset_index()
Summary2.columns = ['Compared to last term, how often do lessons now start without disruptions (e.g., talking or off-task behavior)?  ', 'Count']
Summary2
```

```python
labels = Summary2['Compared to last term, how often do lessons now start without disruptions (e.g., talking or off-task behavior)?  ']
counts = Summary2['Count']
colors = plt.cm.rainbow(np.linspace(0, 1, len(labels)))

# Plot
plt.figure(figsize=(10, 6))
plt.bar(labels, counts)
plt.title('Compared to last term, how often do lessons now start without disruptions (e.g., talking or off-task behavior)?  ')
plt.xlabel('Response')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.yticks(range(0, max(counts)+2, 2))
plt.tight_layout()
plt.show()
```
